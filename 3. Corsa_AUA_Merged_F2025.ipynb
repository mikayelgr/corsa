```python
import pandas as pd

all_courses_df = pd.read_csv("./.localdata/aua__jenza-cleaned-full.csv")
all_courses_df[:2]
```

```python
all_courses_df["Subject Code"].value_counts()
```

```python
all_courses_df[all_courses_df["Subject Code"].str.startswith("GIS")]
```

```python
all_courses_df.loc[all_courses_df["Subject Code"].str.startswith("GIS"), "Subject Code"] = "ESS"
all_courses_df.loc[all_courses_df["Subject Code"] == "ESS", "Course Number"] = True
```

```python
all_courses_df["Subject Code"].value_counts()
```

```python
all_courses_df.loc[all_courses_df["Subject Code"].str.contains("'s Tea")]
```

```python
all_courses_df = all_courses_df.drop(all_courses_df.loc[all_courses_df["Subject Code"].str.contains("'s Tea")].index)
```

```python
geneds_df = pd.read_csv("./.localdata/aua__geneds-cleaned-full.csv")
geneds_df
```

```python
geneds_df["Subject Code"].value_counts()
```

```python
all_courses_df['is_gened'] = False # by default

for i in geneds_df.index:
  gened_entry = geneds_df.loc[i]
  corresponding_offered_courses = all_courses_df.loc[
    (all_courses_df['Course Number'] == gened_entry['Course Number']) &
    (all_courses_df['Subject Code'] == gened_entry['Subject Code'])]
  if len(corresponding_offered_courses) > 0:
    for k in corresponding_offered_courses.index:
      all_courses_df.loc[k, 'is_gened'] = True
      all_courses_df.loc[k, 'Themes'] = gened_entry['Themes']

all_courses_df["is_gened"].fillna(False)
all_courses_df[:5]
```

```python
all_courses_df
```

```python
# Making sure that foundational courses are not marked as gened
all_courses_df.loc[all_courses_df["Subject Code"] == "FND", "is_gened"] = False
```

```python
all_courses_df["Subject Code"].unique()
```

```python
# create a map of subject codes to their full names from all_courses_df["Subject Code"].unique()
subject_code_map = {
    'BSN': 'Nursing',
    'BUS': 'Business',
    'CHSS': 'Humanities and Social Sciences',
    'CS': 'Computer Science',
    'CSE': 'Computer Science and Engineering',
    'DS': 'Data Science',
    'EC': 'English & Communications',
    'ECON': 'Economics',
    'ENGS': 'Engineering',
    'ESS': 'Environmental Science and Sustainability',
    'GIS) and Remote Sensi': 'Geographic Information Systems and Remote Sensing',
    'FND': 'Foundational',
    'HRSJ': 'Human Rights and Social Justice',
    'IESM': 'International and European Studies',
    'IRD': 'Interdisciplinary Research Design',
    'JRN': 'Journalism',
    'LAW': 'Law',
    'PA': 'Public Affairs',
    'PG': 'Politics & Governance',
    'PH': 'Public Health',
    'TEFL': 'Teaching English as a Foreign Language',
}
```

```python
# map courses to their full names
all_courses_df['Major/Concentration Name'] = all_courses_df['Subject Code'].map(subject_code_map)
```

```python
all_courses_df.sample(5)
```

```python
all_courses_df["Session"] = all_courses_df["Session"].str.lower()
all_courses_df.sample(5)
```

```python
all_courses_df.to_csv("./.localdata/aua__all-courses-merged.csv", index=False)
```
